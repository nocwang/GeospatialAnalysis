```python
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import pandas as pd
def get_cities_within_radius(uscities,radius, los_angeles_lat, los_angeles_lon):
    # Use a pandas DataFrame to store the city names and distances
    list_results = []#pd.DataFrame(columns=['city', 'distance'])

    # Loop through all US cities and calculate their distance to Los Angeles
    for city in uscities.itertuples():
        city_lat = city.lat#latitude
        city_lon = city.lng#longitude
        distance = geodesic((los_angeles_lat, los_angeles_lon), (city_lat, city_lon)).miles
        
        # Add the city to the results DataFrame if it's within the radius
        if distance <= radius:
            list_results.append([city.city, distance])

    print(len(list_results))
    return list_results
# Define the radius in miles
radius = 100
uscities=pd.read_csv('uscities.csv')
```

```python
# different online geo tables may be not consistent, 
# e.g., geopy lat/log may be different from  uscities.csv file ( https://simplemaps.com/data/us-cities )
uscities.query("city=='Los Angeles' & state_id=='CA'")[['lat','lng']]
```

```python
# # Use geopy to get the coordinates of Los Angeles
geolocator = Nominatim(user_agent="my-app")
location = geolocator.geocode("Los Angeles, California")
los_angeles_lat = location.latitude
los_angeles_lon = location.longitude
los_angeles_lat, los_angeles_lon
```

# "city=='Los Angeles' & state_id=='CA'"

```python
los_angeles=uscities.query("city=='Los Angeles' & state_id=='CA'")
los_angeles_lat=los_angeles['lat'].values
los_angeles_lon=los_angeles['lng'].values
list_results=get_cities_within_radius(uscities,radius, los_angeles_lat, los_angeles_lon)
```

```python
results=pd.DataFrame(list_results, columns=['city', 'distance'])
# Print the cities within the radius, sorted by distance
print(results.sort_values(by='distance')['city'])
```
